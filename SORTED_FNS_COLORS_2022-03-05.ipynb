```python
def update_scatter(in_fig, dot_vals, dot_colors, night_mode=True):
    """Returns Plotly Scatter figure with prespecified format:
    
    parameters: 
        in_fig: non-formatted scatter figure
        dot_vals: list of values used in x or y axis (for gradient of dot_colors)
        dot_colors: list of lists of val, rgb() pairs specifying color scheme of dot markers
        night_mode: boolean indicating use of off_black or off_white vars in axis and background
        
    assume off_black and off_white vars are declared
        
    for refactor: 
        add actual hex codes instead of off_white and off_black so vars don't need to be declared
        move width and height vars to top and add to documentation - easier to edit later
    """
#     dot_colors=hk_bluered

    text_pos="top center"
    if not night_mode:
        axis_color=off_black
        line_color=off_black
        outline_color=off_black
        font_color=off_black
        bg_color=off_white
    else:
        axis_color=off_white
        line_color=off_white
        outline_color=off_white
        font_color=off_white
        bg_color=off_black 
    fig = in_fig
    fig.update_traces(textposition=text_pos,
                        marker=dict(size=30,
                                    color=dot_vals,
                                    colorscale=dot_colors,
                                     line=dict(color=outline_color, width=2)
                             )
                 )
    #UPDATE COLORS TO: BACKGROUND, PAPER, TEXT, AXIS
    fig.update_layout(
        paper_bgcolor=bg_color, plot_bgcolor=bg_color,
        xaxis=dict(linecolor=axis_color),
        yaxis=dict(linecolor=axis_color),
        font=dict(color=font_color)
    )

    #EDIT GRIDLINES AND AXIS
    fig.update_layout(
        xaxis=dict(showgrid=False,showticklabels=False, showline=True, linewidth=2,
                  range=[0,10]),
        yaxis=dict(showgrid=False, showticklabels=False, showline=True, linewidth=2,
                  range=[0,10])
    )

    #UPDATE FONT
    data_font="Futura"
    title_font= "Futura" #affects the x and y axis title as well

    fig.update_layout(
        font=dict(family=data_font, size=18),
        title_font=dict(family=title_font, size=20),
        xaxis=dict(title_font=dict(family=title_font, size=24)),
        yaxis=dict(title_font=dict(family=title_font, size=24))
    )

    #UPDATE SIZE
    fig.update_layout(
        width=750,
        height=950,
    )
    return fig

```

```python
#test_fn - scatters
for colorscheme in all_colorschemes:
    print("COLORSCHEME")
    print(colorscheme)
    for t_or_f in [True, False]:
        
        update_scatter(fig,y_vals, colorscheme, night_mode=t_or_f).show()
```

```python
def update_bar(in_fig, y_vals, mark_color, night_mode=False):
    """Returns Plotly Bar figure with prespecified format:
    
    parameters: 
        in_fig: non-formatted scatter figure
        y_vals: list of values used in x or y axis (length is used for assigning bar color
        mark_color: string with rgb() or hexcode for color of all bars
        night_mode: boolean indicating use of off_black or off_white vars in axis and background
        
    assume off_black and off_white vars are declared
        
    for refactor: 
        rework logic - need y_vals just to create bar_colors var?
        add actual hex codes instead of off_white and off_black so vars don't need to be declared
        move width and height vars to top and add to documentation - easier to edit later
        fix out_fig to fig to retain consistency
    """
    bar_colors = [mark_color, ]*len(y_vals)

    fig_width = 900
    fig_height = 750
    margin_dict=dict(
                    l=120, 
                    r=120, 
                    t=120, 
                    b=120,
    )
    
    if night_mode:
        axis_color=off_black
        outline_color=off_black
        font_color=off_black
        bg_color=off_white
    else:
        axis_color=off_white
        outline_color=off_white
        font_color=off_white
        bg_color=off_black 

    title_font_dict=dict(family='Futura',color=font_color, size=28)
    axes_font_dict = dict(family='Futura',color=font_color, size=22)
    data_font_dict=dict(family='Futura',color=font_color)
    xaxis_dict=dict(
            title=dict(font=axes_font_dict),
            showgrid=False, showline=True, showticklabels=True,
            linecolor=axis_color, linewidth=3, )
    yaxis_dict=dict(
            title=dict(font=axes_font_dict),
            showgrid=False, showline=True, 
            linecolor=axis_color, linewidth=3, 
            tickmode='array',
            showticklabels=False,
            range=[0, 10])
    out_fig = in_fig
    
    fig.update_traces(marker_color=bar_colors, marker_line_color=axis_color, marker_line_width=1.5)
    fig.update_layout(paper_bgcolor=bg_color, plot_bgcolor=bg_color,
                    xaxis=xaxis_dict, yaxis=yaxis_dict, font=data_font_dict,
                    width=fig_width, height=fig_height, margin=margin_dict )
    fig.update_layout(title_font = title_font_dict)
    return out_fig

```

```python
for color in all_bar_colors:
    print("COLORSCHEME")
    print(color)
    for t_or_f in [True, False]:
        
        update_bar(fig,y_vals, color, night_mode=t_or_f).show()
```
